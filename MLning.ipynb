```python
print("Número original de datapoints\n", data.shape) # Filas x Columnas

# Elimina cualquier valor faltante y remplazarlos hacia la variable de los datos
data = data.dropna()

# Comprueba la nueva forma (shape)
print("Número actualizado de datapoints\n", data.shape) # Filas x Columnas
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# Ajustar el tamaño de la fuente global
plt.rcParams['font.size'] = 15

# Crear una figura y ajustar su tamaño
f = plt.figure(figsize=(8, 4))

# Crear un subplot
ax = f.add_subplot(1, 1, 1)

# Separa los datos en dos grupos: 0 y 1
data_0 = data[data["Outcome"] == 0]
data_1 = data[data["Outcome"] == 1]

# Graficar los histogramas de las dos categorías con colores diferentes
ax.hist(data_0["Outcome"], color='coral', edgecolor='black', linewidth=1.5, label='N', alpha=0.7, bins=10)
ax.hist(data_1["Outcome"], color='royalblue', edgecolor='black', linewidth=1.5, label='Y', alpha=0.7, bins=10)

# Establecer límites en el eje x y ticks
ax.set_xlim([-0.5, 1.5])
ax.set_xticks([0, 1])
ax.set_xticklabels(["N", "Y"])

# Crear el título
ax.set_title("Diabetes Y/N?", fontsize=18, fontweight='bold')

# Establecer etiquetas de los ejes
ax.set_xlabel("Respuesta", fontsize=14)
ax.set_ylabel("# Encuestados", fontsize=14)

# Establecer los límites superior/inferior del eje y
ax.set_ylim([0, 510])

# Ajustar leyenda
ax.legend(title='Outcome', title_fontsize='13')

# Añadir una cuadrícula para mejorar la legibilidad
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Asegurarse de que las cosas sean agradables a la vista
f.tight_layout()

# Mostrar la gráfica
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#  DataFrame
df = pd.read_csv("DB.csv")

# Set up the figure for the histogram
f = plt.figure(figsize=(10, 6))
ax = f.add_subplot(1, 1, 1)

# Define the number of bins
num_bins = 5

# Create the histogram and get the counts and bin edges
counts, bin_edges = np.histogram(data["Age"], bins=num_bins)

# Define a color for each bin
colors = plt.cm.viridis(np.linspace(0, 1, num_bins))  # Use a colormap for variety

# Create the histogram manually to assign different colors
for count, edge, color in zip(counts, bin_edges[:-1], colors):
    ax.bar(edge, count, width=edge - bin_edges[0], color=color, edgecolor='black', linewidth=1, align='edge')

# Configure titles and labels
ax.set_title("Age range of patients", fontsize=16)
ax.set_ylim([0, 10])  # Adjust this limit according to your data
ax.set_xlabel("Age", fontsize=14)
ax.set_ylabel("Count", fontsize=14)

# Adjust the layout
f.tight_layout()

# Show the plot
plt.show()
```
